Если u задано в диапазоне [u, u + delta u], то u_p и u_n будут определяться в некоторых диапазонах.

Минимальный диапазон для u_p и u_n - это [u_p, u_p + 1] и [u_n, u_n + 1]

Поэтому пусть u будет задано в диапазоне [u, u + 2].
