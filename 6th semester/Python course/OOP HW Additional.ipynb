```python
import random as r
```

```python
count = {}
count['K_rho'] = 0
count['K_0(1430) pi'] = 0
count['K(892) pi'] = 0
count['K_omega'] = 0
count['K f_0'] = 0
# count['gamma_K_0'] = 0
```

```python
class particle:
    name = 'K1'
    def __init__(self, name):
        self.name = name
    def display_info(self):
        print('This particle is ' + self.name)
        
        
    def K_rho(self):
        print(self.name + ' decays into K_rho')
        count['K_rho'] += 1
    def K_0_1430_pi(self):
        print(self.name + ' decays into K_0(1430) pi')
        count['K_0(1430) pi'] += 1
    def K_892_pi(self):
        print(self.name + ' decays into K_892 pi')
        count['K(892) pi'] += 1
    def K_omega(self):
        print(self.name + ' decays into K_omega')
        count['K_omega'] += 1
    def K_f_0_1370(self):
        print(self.name + ' decays into K f_0(1370)')
        count['K f_0'] +=1
```

```python
class K1_1270(particle):
    name = 'K1(1270)'
    def display_info(self):
        print('This particle is ' + self.name)
        print('Type: K1(1270)')
    
    def decay(self):
        the_way = r.random()
        #print(the_way)
        if the_way <= 0.42:
            particle.K_rho(self)
        elif the_way <= 0.42 + 0.28:
            particle.K_0_1430_pi(self)
        elif the_way <= 0.42 + 0.28 + 0.16:
            particle.K_892_pi(self)
        elif the_way <= 0.42 + 0.28 + 0.16 + 0.11:
            particle.K_omega(self)
        else:
            particle.K_f_0_1370(self)
```

```python
class K1_1400(particle):
    name = 'K1(1400)'
    def display_info(self):
        print('This particle is ' + self.name)
        print('Type: K1(1400)')
        
    def decay(self):
        the_way = r.random()
        if the_way <= 0.94:
            particle.K_892_pi(self)
        elif the_way <= 0.94 + 0.03:
            particle.K_rho(self)
        elif the_way <= 0.94 + 0.03 + 0.02:
            particle.K_f_0_1370(self)
        else:
            particle.K_omega(self)
```

```python
N_1270 = int(input('N_1270 = ')) # number of particles
N_1400 = int(input('N_1400 = '))

particle_array_1270 = [''] * N_1270
particle_array_1400 = [''] * N_1400

count['K_rho'] = 0
count['K_0(1430) pi'] = 0
count['K(892) pi'] = 0
count['K_omega'] = 0
count['K f_0'] = 0

for i in range(N_1270):
    particle_array_1270[i] = K1_1270('K1(1270)')
    particle_array_1270[i].decay()
    
for i in range(N_1400):
    particle_array_1400[i] = K1_1400('K1(1400)')
    particle_array_1400[i].decay()
    
print('\n', 'statistics: ', str(count))
```
