```python
import random as r
```

```python
class particle:
    name = 'K1(1270)'
    characteriscic = ''
    
    def __init__(self, name, characteriscic): # ПОРЯДОК ВАЖЕН
        self.name = 'aaaaaaaaaaaaaa'
        self.characteriscic = characteriscic
    
    def display_info(self):
        print('This particle is ' + self.name)
        print(self.characteriscic)
```

```python
p3 = particle('NAME', 'CHARACTERISTIC')
p3.display_info()
```

```python
count = {}
count['K_rho'] = 0
count['K_0_1430_pi'] = 0
count['K_892_pi'] = 0
count['K_omega'] = 0
count['K_f_0'] = 0
print(count)
```

```python
class decay_modes(particle):
    def K_rho(self):
        print(self.name + ' decays into K_rho')
        count['K_rho'] += 1
    def K_0_1430_pi(self):
        print(self.name + ' decays into K_0_1430 pi')
        count['K_0_1430_pi'] += 1
    def K_892_pi(self):
        print(self.name + ' decays into K_892 pi')
        count['K_892_pi'] += 1
    def K_omega(self):
        print(self.name + ' decays into K_omega')
        count['K_omega'] += 1
    def K_f_0(self):
        print(self.name + ' decays into K f_0')
        count['K_f_0'] +=1
    def decay(self):
        the_way = r.random()
        #print(the_way)
        if the_way <= 0.42:
            decay_modes.K_rho(self)
        elif the_way <= 0.42 + 0.28:
            decay_modes.K_0_1430_pi(self)
        elif the_way <= 0.42 + 0.28 + 0.16:
            decay_modes.K_892_pi(self)
        elif the_way <= 0.42 + 0.28 + 0.16 + 0.11:
            decay_modes.K_omega(self)
        elif the_way <= 0.42 + 0.28 + 0.16 + 0.11 + 0.03:
            decay_modes.K_f_0(self)
```

```python
p2 = decay_modes('particle 2', 'argument for characteriscic')
p2.display_info()
```

```python
p2.name = "meow"
p2.characteriscic = 'вера пална'
p2.display_info()
```

```python
N = 10000# number of particles

particle_array = [''] * N

count['K_rho'] = 0
count['K_0_1430_pi'] = 0
count['K_892_pi'] = 0
count['K_omega'] = 0
count['K_f_0'] = 0

for i in range(N):
    particle_array[i] = decay_modes(str(i))
    particle_array[i].decay()
    
print('\n', 'statistics: ', str(count))
```
